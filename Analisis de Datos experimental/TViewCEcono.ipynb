 Ejemplo de scraping del calendario económico de TradingView
A continuación, te muestro cómo realizar scraping del calendario económico de TradingView utilizand

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from zoneinfo import ZoneInfo #Nueva importacion para establecer la hora de Peru
```

```python
# Definir el rango de fechas para la consulta
#ZoneInfo es para establecer una zona hora
today = datetime.now(ZoneInfo("America/Lima")).date() 
# Rango de fechas desde hoy hasta 7 días después, con formato ISO
desde = today.isoformat() + 'T00:00:00.000Z'
hasta = (today + timedelta(days=7)).isoformat() + 'T23:59:59.999Z'
```

```python
# Configurar los encabezados y parámetros de la solicitud
headers = {
    'Origin': 'https://www.tradingview.com'
}
params = {
    'from': desde,
    'to': hasta,
    #'countries': 'US,IN'   Puedes ajustar los países según tus necesidades
}
```

```python
# Realizar la solicitud GET al endpoint de TradingView
url = 'https://economic-calendar.tradingview.com/events'
#se pone las variables creadas que se tiene, 
response = requests.get(url, headers=headers, params=params) 
```

```python
if response.status_code == 200:
    data = response.json()
    eventos = data.get('result', [])
    df = pd.DataFrame(eventos)
    
    print(df.columns)  # Verifica los nombres de columnas

    if 'date' in df.columns:
        df['datetime_utc'] = pd.to_datetime(df['date'], utc=True)
        df['datetime_peru'] = df['datetime_utc'].dt.tz_convert('America/Lima')

    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    display(df.head())  # O display(df)
else:
    print(f"Error al obtener los datos: {response.status_code}")
```
