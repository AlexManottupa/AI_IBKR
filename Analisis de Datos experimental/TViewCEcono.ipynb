 Ejemplo de scraping del calendario económico de TradingView
A continuación, te muestro cómo realizar scraping del calendario económico de TradingView utilizand

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Definir el rango de fechas para la consulta
hoy = datetime.utcnow().date()
desde = hoy.isoformat() + 'T00:00:00.000Z'
hasta = (hoy + timedelta(days=7)).isoformat() + 'T23:59:59.999Z'
```

```python
# Configurar los encabezados y parámetros de la solicitud
headers = {
    'Origin': 'https://www.tradingview.com'
}
params = {
    'from': desde,
    'to': hasta,
    'countries': 'US,IN'  # Puedes ajustar los países según tus necesidades
}
```

```python
# Realizar la solicitud GET al endpoint de TradingView
url = 'https://economic-calendar.tradingview.com/events'
response = requests.get(url, headers=headers, params=params)
```

```python
# Verificar el estado de la respuesta
if response.status_code == 200:
    data = response.json()
    eventos = data.get('result', [])
    df = pd.DataFrame(eventos)
    display(df.head())
else:
    print(f"Error al obtener los datos: {response.status_code}")
```
