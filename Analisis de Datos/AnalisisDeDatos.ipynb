Importaciones de las librerias a utilizar


```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python

# Cargar los datos desde el archivo Excel
excel_path = "datos_financieros.xlsx"
df = pd.read_excel(excel_path)
```

```python
# Convertir la columna "Hora (UTC)" a un índice datetime
df['Hora (UTC)'] = pd.to_datetime(df['Hora (UTC)'], format='%H:%M')
df.set_index('Hora (UTC)', inplace=True)
```

```python
# Renombrar las columnas para que mplfinance las reconozca
df.rename(columns={"Apertura": "Open", "Máximo": "High", "Mínimo": "Low", "Cierre": "Close"}, inplace=True)
```

```python
# Definir los colores de las velas
mc = mpf.make_marketcolors(up='green', down='blue', inherit=True)

# Estilo de las velas
mpf_style = mpf.make_mpf_style(base_mpf_style="charles", marketcolors=mc, gridstyle="--", y_on_right=True)
```

```python

# Graficar las velas
mpf.plot(df, type="candle", style=mpf_style, title="GBP/USD - 8 de noviembre de 2024", ylabel="Precio (USD)",
         volume=False, figsize=(10, 6), datetime_format='%H:%M', )

# Mostrar el gráfico
plt.show()
```
