```python
import undetected_chromedriver as uc
from bs4 import BeautifulSoup
import time
from datetime import datetime, timedelta
```

```python
# Obtener fecha de hoy y ayer
today = datetime.now().strftime("%b %d, %Y")  # Formato "Mar 15, 2025"
yesterday = (datetime.now() - timedelta(days=1)).strftime("%b %d, %Y")
```

```python
# Iniciar el navegador
driver = uc.Chrome(headless=True)
driver.get("https://www.investing.com/economic-calendar/")
```

```python
# Esperar que la página cargue completamente
time.sleep(5)
```

```python
# Obtener HTML de la página
html_content = driver.page_source
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# Buscar eventos del calendario
events = soup.find_all("tr", class_="js-event-item")
```

```python

# Lista para almacenar datos
data = []
```

```python
for event in events:
    # Extraer datos del evento
    date = event.get("data-event-datetime", "")  # Fecha del evento
    time_event = event.find("td", class_="first left time")
    country = event.find("td", class_="left flagCur noWrap")
    event_name = event.find("td", class_="left event")
    impact = event.find("td", class_="sentiment")
    previous = event.find("td", class_="bold previous")
    actual = event.find("td", class_="bold actual")
    forecast = event.find("td", class_="bold forecast")
    
    # Validar que los elementos existen antes de extraer texto
    if date and (yesterday in date or today in date):
        data.append([
            time_event.text.strip() if time_event else 'N/A',
            country.text.strip() if country else 'N/A',
            event_name.text.strip() if event_name else 'N/A',
            impact.text.strip() if impact else 'N/A',
            previous.text.strip() if previous else 'N/A',
            actual.text.strip() if actual else 'N/A',
            forecast.text.strip() if forecast else 'N/A'
        ])

# Cerrar navegador
driver.quit()
```

```python
import pandas as pd
```

```python
# Crear DataFrame con los eventos
df = pd.DataFrame(data, columns=["Hora", "País", "Evento", "Impacto", "Previo", "Actual", "Pronóstico"])
```
