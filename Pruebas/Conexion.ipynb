Conexion with api of interactive brokers


```python
ib = IB()
ib.connect('127.0.0.1', 7497, clientId=1)
```

```python
contract = Forex('GBPUSD')
bars = ib.reqHistoricalData(
    contract, endDateTime='', durationStr='2 D',
    barSizeSetting='1 hour', whatToShow='MIDPOINT', useRTH=True)
```

```python
# convert to pandas dataframe
df = util.df(bars)
print(df)
```

#Graficamos el plano, damos los valores para que construya el plano

```python
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['close'], label='Close Price')

# Añadir títulos y etiquetas
plt.title('GBP/USD - precio MIDPOINT (1 D)') # sELECCIONAMOS LA DIVISA
plt.xlabel('Fecha')
plt.ylabel('Cierre del precio en esa hora respectiva')
plt.xticks(rotation=45)
plt.legend()


# Despues de definir de como va esta mi ploteo lo muestro finalmente
plt.show()
```
